```python
import pandas as pd

# load the data
data = pd.read_csv('../data/processed/combined_data.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation matrix')
plt.show()
```

```python
# Feature histograms
data.hist(figsize=(15, 10), bins=20)
plt.tight_layout()
plt.show()
```

```python
# Class-wise Distribution
sns.boxplot(x='Cooler condition %', y='Temperature 1 (C)', data=data)
plt.title("Temperature 1 vs Cooler Condition")
plt.show()
```
